來源 : Edge

## 1.  NC-file.schema.json 

**作用**：  
定義整支 NC 檔案的基本資訊，例如來源、機台、座標模式與製程類型。

| 欄位 | 是否必填 | 說明 |
|---|---|---|
|  program_id  | 是 | NC 程式的識別碼 |
|  file_name  | 是 | 檔名 |
|  source_format  | 是 | 檔案格式，限定為  Siemens MPF  |
|  source_variant  | 是 | 檔案版本，例如  original 、 modified  |
|  cam_system  | 是 | 產生這份程式的 CAM 系統 |
|  source_cam_file  | 否 | 上游 CAM 檔名或路徑 |
|  machine  | 是 | 機台名稱或型號 |
|  machine_type  | 是 | 機台類型 |
|  created_date  | 是 | 建立日期 |
|  created_time  | 是 | 建立時間 |
|  unit_system  | 是 | 單位系統，這裡限定為  mm  |
|  coordinate_plane  | 是 | 加工平面，例如  G17  |
|  coordinate_mode  | 是 | 座標模式，例如  G90  |
|  work_offset  | 是 | 工件座標系，例如  G54  |
|  path_control_mode  | 是 | 路徑控制模式，例如  G64  |
|  transform_mode  | 是 | 姿態／座標轉換模式，例如  TRAORI(2)  |
|  home_position_used  | 是 | 是否使用 home position |
|  has_laser_process  | 是 | 是否包含雷射製程 |
|  has_powder_supply  | 是 | 是否包含送粉控制 |
